```python
import opendatasets as od
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = 'https://www.kaggle.com/datasets/yasserh/breast-cancer-dataset'
od.download(data)
```

```python
import numpy as np
import pandas as pd
```

```python
cancer_data = pd.read_csv('breast-cancer.csv')
cancer_data.head()
```

```python
Y = cancer_data['diagnosis']
X = cancer_data.drop(columns='diagnosis')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

Finding the **Accuracy Score**

```python
model = RandomForestClassifier()
model.fit(X_train_std,Y_train)
Y_predict = model.predict(X_test_std)
print("The accuracy score is: ", accuracy_score(Y_test,Y_predict))
```

```python
pred_values = pd.DataFrame({'Actual value' : Y_test , 'Predicted value': Y_predict})
pred_values.head(n=10)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf = confusion_matrix(Y_test,Y_predict)
print(cnf)
```
